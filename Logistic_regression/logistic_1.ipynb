```python
from  sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples = 100, n_features = 2 , n_informative = 1 , n_redundant=0,
                          n_classes = 2 , n_clusters_per_class =1 , random_state = 41, hypercube = False , class_sep = 10)

#Generate 100 points with 2 features, only 1 feature truly matters, no duplicate features, 2 classes, one cluster per class, very large separation between classes, giving an easy binary classification problem.
```

```python
import matplotlib.pyplot as plt
X
```

```python
plt.figure(figsize= (10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap = "winter" , s= 100)
```

```python
def step(z):
    return 1 if z > 0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis =1)
    weights = np.ones (X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr * (y[j]- y_hat)*X[j]
    return weights[0], weights[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
# np.dot(np.insert(X[41], 0, 1), [1,1,1])
'''First we make a new column for the intercept in X and set is as one using np.ones. Secondly, we do dot product with the weights and the result obtained is positive or else it is negative that means our model prediction. If its positive assign it 1  if its negative assign it 0

Then we compare with our dataset what is its actual value 

If our dataset says it's  positive but our model predicts its negative the add the product of learning rate with the weights and then calculate the new weights
'''
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m *x_input + b 
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input , y_input , color = 'red', linewidth = 3)
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter', s= 100)
plt.ylim(-3,2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
plt.figure(figsize = (10,6))
plt.plot(x_input, y_input, color ='red', linewidth = 3)
plt.plot(x_input1, y_input1, color ='black', linewidth = 3)
plt.scatter(X[:,0],X[:,1], c= y, cmap= 'winter', s = 100)
plt.ylim(-3,2)
# line width makes the line thicker
#plt.ylim() sets the visibility range for y - axis...
# sep attribute separates both classes afar
```
