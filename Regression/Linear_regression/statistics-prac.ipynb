```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
reg = LinearRegression()
```

```python
data = pd.read_csv("advertising.csv")
print(data.iloc[0:202,0:2])
```

```python
plt.scatter(data["TV"], data["Sales"])
```

```python
X = data[["TV"]]
Y = data["Sales"]
reg.fit(X,Y)
```

```python
slope = reg.coef_
print(slope[0])
```

```python
intercept = reg.intercept_
intercept
```

```python
print(f"The LInear regression eqn is : Y = {slope[0]} X + {intercept}")
```

```python
prediction = reg.predict(X)
prediction
```

```python
variation = r2_score(Y,prediction)
variation
```

```python
error = mean_squared_error(prediction, Y)
error
```

```python
plt.scatter(X, Y, label="Actual Data")
plt.plot(X, prediction, color = "purple" , label="Best Fit Line")
```
