```python
import numpy as np
import pandas as pd
from sklearn import linear_model
reg = linear_model.LinearRegression()
data = {
    "X" : np.array([1,2,3,4]),
    "Y" : np.array([3,5,11,19])
}

df = pd.DataFrame(data)
print (df)
x = df[["X"]].values
y = df["Y"].values
reg.fit(x,y)
print("Coefficient (slope):", reg.coef_)
print("Intercept:", reg.intercept_)

```
