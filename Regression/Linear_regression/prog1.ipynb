```python
df = df.dropna()
df
```

```python
df["Extracurricular Activities"] = df["Extracurricular Activities"].map({"Yes": 1, "No": 0})
df["Extracurricular Activities"]
print(df.nunique())
df["Performance Index"].unique()#gives unique valuesof the specific column
```

```python
df.isna()
```

```python
df.drop(["Sleep Hours"], axis=1,inplace=True)
df
```

```python
# List of numeric columns to check
numeric_cols = ['Hours Studied', 'Previous Scores', 
                'Sample Question Papers Practiced', 'Performance Index']

# Loop through each column
for col in numeric_cols:
    # Calculate quartiles and IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Detect outliers (print how many)
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col} outliers:", len(outliers))

    # Remove outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python


pairs = [
    ('Hours Studied', 'Performance Index'),
    ('Previous Scores', 'Performance Index'),
    ('Sample Question Papers Practiced', 'Performance Index')
]

for x, y in pairs:
    plt.figure()
    plt.scatter(df[x], df[y], alpha=0.6) 
    plt.title(f'{x} vs {y}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

y = df['Performance Index'].values
X = df.drop('Performance Index', axis=1)
X.head()
```

```python
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2,     # 20% test data
    random_state=42    # for reproducibility
)

lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
r2 = r2_score(y_test,y_pred)
r2
```
