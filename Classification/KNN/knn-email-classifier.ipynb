In raw word based classification, we must transform the words into numbers so that machine could understand and train the model accordingly

So, we use tf-idf vectorizer technique to calculate the word frequency and convert it into number so that it is understandable by machine to implement knn.

TF : Term Frequency
IDF : Inverse Document Frequency

TF = word frequency /Total words
IDF = log(Total documents/documents containing the word)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier

# Sample dataset
emails = [
    "Win money now",
    "Limited offer discount",
    "Meeting schedule tomorrow",
    "Project discussion today",
    "Claim your free prize"
]

labels = [
    "spam",
    "spam",
    "not spam",
    "not spam",
    "spam"
]

# Convert text → numerical vectors
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(emails)
print(X)
# Train KNN model
k = 3
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X, labels)

# Test email
test_email = ["Free money offer"]
test_vector = vectorizer.transform(test_email)

prediction = model.predict(test_vector)
print(prediction)
```
