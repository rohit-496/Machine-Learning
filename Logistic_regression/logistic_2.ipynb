```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python

x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

In this perceptron trick, when the points are classified right it stops ,and it only changes the orientation of thwe line whenever the point is misplaced such that it doesnt create equilibrium or symmetry between the data points.

Because when the step fucntion is called it gives 1 or 0 which can change the whole learning rate term to 0 and no change is brought to the orientation of the line and the weights are same as the old weights

```python

m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

This is the logistic regression algorithm from the Scikit-learn library which gives the optimal solution of the logistic regression by dividing the sapce and classifying the data points according to the symmetry such that they both are equally separated

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```
