```python

from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
```

```python
X,y = fetch_california_housing(return_X_y=True)
```

```python
housing = fetch_california_housing()
```

```python
df = pd.DataFrame(X,y)
df
```

```python
df = pd.DataFrame(
    housing.data,
    columns=housing.feature_names
)
```

```python
df["price"] = housing.target
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [("lr", lr),("dt",dt),("svr", svr)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], X,y,cv = 5, scoring = "r2")
    print(estimator[0],np.mean(scores))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators = estimators)
scores = cross_val_score(estimator[1], X,y,cv = 10, scoring = "r2")
print("Voting Regressor",np.mean(scores))
```

```python

vr = VotingRegressor(estimators,weights=[0.7,0.3,0.0])
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print(np.round(np.mean(scores),2))
```
